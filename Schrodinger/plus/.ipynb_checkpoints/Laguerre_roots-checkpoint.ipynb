```julia
using IntervalArithmetic, Combinatorics, Polynomials, PolynomialRoots, Serialization, LinearAlgebra
```

```julia
# all computations are bounded rigorously via Interval Arithmetic
setprecision(Interval , 8192)

# computes enclosure for rationals
intervalR(x::Rational) = @interval(x)
```

```julia
# Gaussian-Laguerre quadrature exact for polynomials of degree 2n-1
n = big(1002)
# Laguerre polynomial of order n and index 0
Lₙ = Polynomial([(-1)^k*intervalR(binomial(n,k)//factorial(k)) for k=0:n]);
# fast computation in arbitrary precisions of the roots of Lₙ
X = real.(PolynomialRoots.roots(mid.([(-1)^k*intervalR(binomial(n,k)//factorial(k)) for k=0:n])));
```

```julia
# rigorous enclosure of the roots of Lₙ via ϵ-inflation
ϵ = @interval(2^(-4096-2048))
Xl = X .- ϵ
Xu = X .+ ϵ
diff = Lₙ.(Xl) -Lₙ.(Xu)
# # check enclosure via intermediate value Thm and fundamental Thm of Algebra (if true)
println(!any(0 .∈ diff))
Xrig = [interval(Xl[i].lo, Xu[i].hi) for i=1:n];
```

```julia
# computation of Lₘ where m = n+1
Lₘ = Polynomial([(-1)^k*intervalR(binomial(n+1,k)//factorial(k)) for k=0:n+1]);
# weights for the Gaussian-Legendre quadrature
W = Xrig ./((Lₘ.(Xrig)).^2)./((n+1)^2);
```

```julia
# quadrature is exact for quadruple product of product of degree n÷2-1
V = zeros(Interval{BigFloat}, (n, n ÷ 2))
Yrig = Xrig ./3
for k = 0:(n÷2)-1
    println(k)
    Lₖ = Polynomial([(-1)^i*intervalR(binomial(k,i)//factorial(i)) for i=0:k])
    V[:,k+1] = Lₖ.(Yrig)
end
```

```julia
# regularized Vandermonde matrix for quadruple integral product
V̄ = Diagonal(W.^(1/4))*V;
```

```julia
# quick test ∫(1-z)^4 exp(-3z) dz
println(11//81 ∈ sum(V̄[:,2].^4)/3)
println(diam(sum(V̄[:,2].^4)/3))
```

```julia
# save data
serialize("Xrig", Xrig)
serialize("Wrig", W)
serialize("Vandermonde", V)
serialize("VandermondeR", V̄)
```
