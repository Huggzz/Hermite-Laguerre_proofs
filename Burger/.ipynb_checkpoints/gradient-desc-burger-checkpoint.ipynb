```julia
using IntervalArithmetic, LinearAlgebra, Serialization, ClassicalOrthogonalPolynomials, Plots
```

```julia
# refularised Vandermonde matrix
V̄ = Float64.(mid.(deserialize("V4r")));
DV̄ = Float64.(mid.(deserialize("DV4r")));
```

```julia
# generator + 1/2 *I
N = 500
L = Diagonal(-collect(1:N+1))+0.75*I
```

```julia
# computes the gradient
function Grad(U)
    W = V̄*U
    DW = DV̄*U
    G = L*U - (V̄'*(W.^2 .*DW))
#     println(norm(G))
    return L*G - 2*V̄'*((W.*DW).*(V̄*G)) - V̄'*((W.^2).*(DV̄*G))
end
```

```julia
γ = 2^(-15)./collect(1:501)
U = zeros(N+1)
U[1] = 1.57
for i=0:200000
    U[2:end] -= (γ.*Grad(U))[2:end]
    if i % 10000 == 0
        println(norm(L*U - (V̄'*((V̄*U).^2 .*(DV̄*U)))))
    end
end
```

```julia
γ = 2^(-12)./collect(1:501)
for i=0:500000
    U -= γ.*Grad(U)
    if i % 10000 == 0
        println(norm(L*U - (V̄'*((V̄*U).^2 .*(DV̄*U)))))
    end
end
```

```julia
function buildnorm_sol(u, xloc)
    K = length(xloc)
    sol = BigFloat.(zeros(K))
    N = BigInt(length(u)-1)
    sqloc = xloc.^2/4
    for i = 0:N
        sol += u[i+1]*laguerrel.(i, sqloc)
#         println(sol[1])
    end
    return sol.*exp.(-sqloc/2)
end
```

```julia
xloc = collect(0:700)/100
sol = buildnorm_sol(U, xloc)
```

```julia
plot(xloc, sol)
```
