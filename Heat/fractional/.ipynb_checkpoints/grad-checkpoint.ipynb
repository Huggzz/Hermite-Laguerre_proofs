```julia
using Serialization, IntervalArithmetic, LinearAlgebra, Polynomials
```

```julia
V̄ = mid.(deserialize("Vr"));
```

```julia
n = 1200
L = Diagonal(big.((-0.5).+(0:n)))

function grad(u)
    U = cbrt.(V̄*u)
    w = L*u + V̄'*(U.^5)
    return L*w + (5//3)*V̄'*((U.^2).* (V̄*w))
end
```

```julia
u = zeros(n+1)
u[1] = 2.0
# u[2] = 0.5
```

```julia
γ = 8*10^(-5)
for i = 0:100000000
    if i % 2000 == 0
        println(norm(L*u + V̄'*(cbrt.(V̄*u).^5)))
    end
    u -= γ*inv(L+3.0*I)*grad(u)
    # u[1] = 1.0
end
```

```julia
w = L*u - V̄'*(V̄*u).^4
```

```julia
U = cbrt.(V̄*big.(u))
PFu = L*u + V̄'*(U.^5)
```

```julia
Gu = L + (5//3)*V̄'*((U.^2) .* V̄)
```

```julia
ũ = u - Gu\PFu
```
